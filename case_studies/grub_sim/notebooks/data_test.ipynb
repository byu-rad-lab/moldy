```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
train_grub_inputdata = np.load('/home/daniel/catkin_ws/src/moldy/case_studies/grub_sim/data/good_data/train_inputdata.npy')
train_grub_outputdata = np.load('/home/daniel/catkin_ws/src/moldy/case_studies/grub_sim/data/good_data/train_outputdata.npy')
```

```python
print(train_grub_inputdata.shape)
print(train_grub_outputdata.shape)
```

```python
print(train_grub_inputdata[0])
print(train_grub_outputdata[0])

inputs = train_grub_inputdata
outputs = train_grub_outputdata
```

```python
# Check the min/max and standard deviation of the inputs and outputs
# set np print to print only 5 decimals
print("Inputs")
print(f"p0: min: {np.min(inputs[:,0]):.8f}, max: {np.max(inputs[:,0]):.8f}, mean: {np.mean(inputs[:,0]):.8f} std: {np.std(inputs[:,0]):.8f}")
print(f"p1: min: {np.min(inputs[:,1]):.8f}, max: {np.max(inputs[:,1]):.8f}, mean: {np.mean(inputs[:,1]):.8f} std: {np.std(inputs[:,1]):.8f}")
print(f"p2: min: {np.min(inputs[:,2]):.8f}, max: {np.max(inputs[:,2]):.8f}, mean: {np.mean(inputs[:,2]):.8f} std: {np.std(inputs[:,2]):.8f}")
print(f"p3: min: {np.min(inputs[:,3]):.8f}, max: {np.max(inputs[:,3]):.8f}, mean: {np.mean(inputs[:,3]):.8f} std: {np.std(inputs[:,3]):.8f}")
print(f"ud: min: {np.min(inputs[:,4]):.8f}, max: {np.max(inputs[:,4]):.8f}, mean: {np.mean(inputs[:,4]):.8f} std: {np.std(inputs[:,4]):.8f}")
print(f"vd: min: {np.min(inputs[:,5]):.8f}, max: {np.max(inputs[:,5]):.8f}, mean: {np.mean(inputs[:,5]):.8f} std: {np.std(inputs[:,5]):.8f}")
print(f"u : min: {np.min(inputs[:,6]):.8f}, max: {np.max(inputs[:,6]):.8f}, mean: {np.mean(inputs[:,6]):.8f} std: {np.std(inputs[:,6]):.8f}")
print(f"v : min: {np.min(inputs[:,7]):.8f}, max: {np.max(inputs[:,7]):.8f}, mean: {np.mean(inputs[:,7]):.8f} std: {np.std(inputs[:,7]):.8f}")
```

```python
# check min/max of the outputs
print("Outputs")
print(f"p0: min: {np.min(outputs[:,0]):.8f}, max: {np.max(outputs[:,0]):.8f}, mean: {np.mean(outputs[:,0]):.8f} std: {np.std(outputs[:,0]):.8f}")
print(f"p1: min: {np.min(outputs[:,1]):.8f}, max: {np.max(outputs[:,1]):.8f}, mean: {np.mean(outputs[:,1]):.8f} std: {np.std(outputs[:,1]):.8f}")
print(f"p2: min: {np.min(outputs[:,2]):.8f}, max: {np.max(outputs[:,2]):.8f}, mean: {np.mean(outputs[:,2]):.8f} std: {np.std(outputs[:,2]):.8f}")
print(f"p3: min: {np.min(outputs[:,3]):.8f}, max: {np.max(outputs[:,3]):.8f}, mean: {np.mean(outputs[:,3]):.8f} std: {np.std(outputs[:,3]):.8f}")
print(f"ud: min: {np.min(outputs[:,4]):.8f}, max: {np.max(outputs[:,4]):.8f}, mean: {np.mean(outputs[:,4]):.8f} std: {np.std(outputs[:,4]):.8f}")
print(f"vd: min: {np.min(outputs[:,5]):.8f}, max: {np.max(outputs[:,5]):.8f}, mean: {np.mean(outputs[:,5]):.8f} std: {np.std(outputs[:,5]):.8f}")
print(f"u : min: {np.min(outputs[:,6]):.8f}, max: {np.max(outputs[:,6]):.8f}, mean: {np.mean(outputs[:,6]):.8f} std: {np.std(outputs[:,6]):.8f}")
print(f"v : min: {np.min(outputs[:,7]):.8f}, max: {np.max(outputs[:,7]):.8f}, mean: {np.mean(outputs[:,7]):.8f} std: {np.std(outputs[:,7]):.8f}")
```

```python
# create 3 subplots where the first contains pressure comparison, second contains velocity comparison, and third contains position comparison
fig, axs = plt.subplots(3, 1, figsize=(10,10))
axs[0].plot(inputs[:,0], 'r-', markersize=1, label='p0_norm')
axs[0].plot(inputs[:,1], 'g-', markersize=1, label='p1_norm')
axs[0].plot(inputs[:,2], 'm-', markersize=1, label='p2_norm')
axs[0].plot(inputs[:,3], 'k-', markersize=1, label='p3_norm')
axs[0].set_title('Pressures')
axs[0].legend()

axs[1].plot(inputs[:,4], 'r-', markersize=1, label='udot_norm')
axs[1].plot(inputs[:,5], 'g-', markersize=1, label='vdot_norm')
axs[1].set_title('Velocity')
axs[1].legend()

axs[2].plot(inputs[:,6], 'r-', markersize=1, label='u_norm')
axs[2].plot(inputs[:,7], 'g-', markersize=1, label='v_norm')
axs[2].set_title('Position')
axs[2].legend()
plt.show()
```
